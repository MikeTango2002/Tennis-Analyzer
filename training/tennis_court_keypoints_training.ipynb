# Start Code


```python
import torch
from torch.utils.data import Dataset, DataLoader
from torchvision import models, transforms
from torchvision.models import ResNet50_Weights

import json
import cv2
import numpy as np
```

DATASET

La classe Dataset è utilizzata per rappresentare un set di dati. È una classe base astratta che deve essere ereditata e i metodi __len__ e __getitem__ devono essere implementati. Questi metodi permettono di:

__len__: Restituire la dimensione del dataset, cioè il numero totale di campioni disponibili.

__getitem__: Permettere di accedere ai campioni del dataset. Dato un indice, restituisce un campione (e la sua etichetta associata, se presente).

DATALOADER 

Il DataLoader è utilizzato per iterare attraverso il Dataset. Fornisce un metodo semplice per caricare i dati in mini-batch, mescolare i dati, e gestire il parallelismo. Alcune delle funzionalità chiave del DataLoader sono:

Batching: Raggruppare i campioni in mini-batch per l'addestramento.
Shuffling: Mescolare i dati per evitare che l'ordine dei dati influisca sull'addestramento.
Parallelismo: Utilizzare più processi per caricare i dati in parallelo, migliorando l'efficienza.

MODELS

La libreria torchvision.models contiene implementazioni pre-addestrate di molte reti neurali convoluzionali (CNN) popolari. Questi modelli sono stati addestrati su grandi dataset, come ImageNet, e possono essere utilizzati per una varietà di compiti di visione artificiale, come il riconoscimento di oggetti, la classificazione delle immagini, l'estrazione di caratteristiche, ecc. Alcuni esempi di modelli inclusi sono ResNet, VGG, Inception, e molti altri.

TRANSFORMS

La libreria torchvision.transforms fornisce strumenti per trasformare e manipolare le immagini. Queste trasformazioni possono essere utilizzate per pre-elaborare le immagini prima di passarle al modello, e includono operazioni comuni come ridimensionamento, ritaglio, rotazione, normalizzazione e conversione del formato.

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(device)
```

# Create Torch Dataset

```python
class KeypointsDataset(Dataset):
    def __init__(self, img_dir, data_file):
        self.img_dir = img_dir
        with open(data_file, 'r') as f:
            self.data = json.load(f)

        self.transforms = transforms.Compose([ #Lista delle trasformazioni da applicare all'immagine
            transforms.ToPILImage(), #Trasforma l'immagine in un oggetto PIL (Python Imaging Library)
            transforms.Resize((224, 224)), #Ridimensiona l'immagine a 224x224 pixel
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]) # sono le medie e le deviazioni standard dei canali di colore rosso, verde e blu (RGB) calcolate su un grande dataset di immagini, comunemente il dataset ImageNet.
        ])

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        item = self.data[idx]
        img = cv2.imread(f"{self.img_dir}/{item['id']}.png")
        h, w = img.shape[:2] #Prende le prime due dimensioni dell'immagine (la terza è il numero di canali di colore)
        
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        img = self.transforms(img)
        
        keypoints = np.array(item['kps']).flatten() #Siccome sono salvati in una lista, li converto in un array numpy 1d
        keypoints = keypoints.astype(np.float32)    #flatten serve a trasformare l'array in un array 1d

        #Le coordinate dei keypoints si riferiscono all'immagine originale, quindi le normalizzo per adattarle all'immagine ridimensionata

        keypoints[::2] *= 224.0 / w #Normalizza le coordinate x (width)
        keypoints[1::2] *= 224.0 / h #Normalizza le coordinate y (height)

        return img, keypoints
```

```python
train_dataset = KeypointsDataset("data/images", "data/data_train.json")
val_dataset = KeypointsDataset("data/images", "data/data_val.json")

train_loader = DataLoader(train_dataset, batch_size=8, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=8, shuffle=True)
```

# Create the Model

```python
model = models.resnet50(weights=ResNet50_Weights.DEFAULT) #Carica il modello preaddestrato ResNet-50

#Applico un fine-tuning del modello, sostituendo l'ultimo layer con un layer completamente connesso 
#con 28 output (14 keypoints * 2 coordinate)

model.fc = torch.nn.Linear(model.fc.in_features, 14 * 2) #Rimpiazza l'ultimo layer 
```

```python
model = model.to(device) #Sposta il modello sulla GPU se disponibile
```

# Train the Model

```python
criterion = torch.nn.MSELoss() #Funzione di loss Mean Squared Error
optimizer = torch.optim.Adam(model.parameters(), lr=1e-4) #Ottimizzatore Adam
```

```python
epochs = 20
for epoch in range(epochs):
    for i, (imgs, keypoints) in enumerate(train_loader):
        imgs = imgs.to(device)
        keypoints = keypoints.to(device)

        #Prima di ogni forward pass, si azzerano i gradienti accumulati dai parametri del modello. 
        #Questo è necessario perché per default, PyTorch accumula (somma) i gradienti su chiamate 
        #successive a backward(). Se non si azzerano i gradienti, i nuovi gradienti verranno sommati 
        #ai gradienti precedenti, portando a risultati errati nell'aggiornamento dei pesi.
        optimizer.zero_grad() 


        ouputs = model(imgs) #Forward pass
        loss = criterion(ouputs, keypoints) #Calcolo della loss
        loss.backward() #Calcolo dei gradienti
        optimizer.step() #Aggiornamento dei pesi

        if i % 10 == 0:
            print(f"Epoch {epoch}, Batch {i}, Loss {loss.item()}")
```
